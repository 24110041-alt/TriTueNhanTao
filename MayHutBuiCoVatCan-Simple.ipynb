# Vacuum Cleaner Agent — PEAS Model
 P — Performance Measure
- Làm sạch càng nhiều càng tốt
- Di chuyển hợp lý
- Giảm số bước di chuyển
 E — Environment
- Môi trường dạng lưới 3x3
- Ô sạch hoặc có bụi
- Robot hoạt động trong phòng
 A — Actuators
- Move Up
- Move Down
- Move Left
- Move Right
- Suck
 S — Sensors
- Cảm biến bụi
- Cảm biến vị trí

# Rule Set — Simple Reflex Vacuum Agent With Obstacles

- Ma trận m x n
- 0 → không có bụi
- 1 → có bụi
- 2 → vị trí robot
- X → vật cản

Rule 1 — Hút bụi

IF state_value = 1
THEN:
- hút bụi
- state_value = 0

Rule 2 — Di chuyển

Robot chỉ đi các hướng hợp lệ và không chứa vật cản.

Rule 3 — Kiểm tra vật cản

IF ô phía trước = X
THEN:
- không di chuyển vào ô đó
- chọn hướng khác

Rule 4 — Các hướng hợp lệ

Góc trên trái (0,0)
Action = RANDOM(DOWN, RIGHT)

Góc trên phải (0,n-1)
Action = RANDOM(DOWN, LEFT)

Góc dưới trái (m-1,0)
Action = RANDOM(UP, RIGHT)

Góc dưới phải (m-1,n-1)
Action = RANDOM(UP, LEFT)

Cạnh trên
Action = RANDOM(DOWN, LEFT, RIGHT)

Cạnh dưới
Action = RANDOM(UP, LEFT, RIGHT)

Cạnh trái
Action = RANDOM(UP, DOWN, RIGHT)

Cạnh phải
Action = RANDOM(UP, DOWN, LEFT)

Ô giữa
Action = RANDOM(UP, DOWN, LEFT, RIGHT)

Lưu ý:
- Các hướng có vật cản sẽ bị loại bỏ trước khi random.

```python
import random
import time

# Kích thước ma trận
m = 5
n = 5

# Tạo môi trường
# 0 = sạch
# 1 = bụi
# -1 = vật cản
room = [[random.choice([0,1]) for _ in range(n)] for _ in range(m)]

# Tạo vật cản ngẫu nhiên
num_obstacles = 4

for _ in range(num_obstacles):

    ox = random.randint(0, m - 1)
    oy = random.randint(0, n - 1)

    room[ox][oy] = -1

# Vị trí robot ban đầu
while True:

    x = random.randint(0, m - 1)
    y = random.randint(0, n - 1)

    # Robot không được đứng vào vật cản
    if room[x][y] != -1:
        break


# Hàm in ma trận
def print_room():

    for i in range(m):

        for j in range(n):

            # Robot
            if i == x and j == y:
                print("2", end=" ")

            # Vật cản
            elif room[i][j] == -1:
                print("X", end=" ")

            else:
                print(room[i][j], end=" ")

        print()

    print()


# Kiểm tra còn bụi không
def has_dust():

    for row in room:

        if 1 in row:
            return True

    return False


# Tìm hướng đi hợp lệ
def possible_moves(x, y):

    moves = []

    # UP
    if x > 0 and room[x - 1][y] != -1:
        moves.append("U")

    # DOWN
    if x < m - 1 and room[x + 1][y] != -1:
        moves.append("D")

    # LEFT
    if y > 0 and room[x][y - 1] != -1:
        moves.append("L")

    # RIGHT
    if y < n - 1 and room[x][y + 1] != -1:
        moves.append("R")

    return moves


# Bắt đầu mô phỏng
step = 1

while has_dust():

    print(f"===== STEP {step} =====")

    print_room()

    # Nếu ô hiện tại có bụi
    if room[x][y] == 1:

        print(f"Robot hút bụi tại ({x},{y})")

        room[x][y] = 0

    # Tìm hướng đi hợp lệ
    moves = possible_moves(x, y)

    print("Possible moves:", moves)

    # Nếu robot bị chặn hết đường
    if len(moves) == 0:

        print("Robot bị chặn bởi vật cản!")

        break

    # Chọn hướng ngẫu nhiên
    action = random.choice(moves)

    print("Robot move:", action)

    # Cập nhật vị trí
    if action == "U":
        x -= 1

    elif action == "D":
        x += 1

    elif action == "L":
        y -= 1

    elif action == "R":
        y += 1

    step += 1

    time.sleep(1)

# Kết thúc
print("===== HOÀN THÀNH =====")

print("Ma trận cuối cùng:")

print_room()
```
