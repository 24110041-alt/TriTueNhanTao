IF bắt đầu thuật toán

    node ← NODE(problem.INITIAL)

IF problem.GOAL-TEST(node.STATE)
THEN
    return SOLUTION(node)

Rule 3 — Tạo frontier

IF node được tạo
THEN
    frontier ← FIFO-QUEUE()
    frontier.INSERT(node)

Rule 4 — Khởi tạo reached

IF frontier đã tạo
THEN
    reached ← ∅

Rule 5 — Kiểm tra frontier

IF frontier ≠ EMPTY
THEN
    tiếp tục thuật toán BFS

Rule 6 — Lấy node đầu tiên

IF frontier ≠ EMPTY
THEN
    node ← frontier.REMOVE()

Rule 7 — Đưa state vào reached

IF node được lấy ra khỏi frontier
THEN
    reached ← reached ∪ {node.STATE}

Rule 8 — Kiểm tra goal

IF problem.GOAL-TEST(node.STATE)
THEN
    return SOLUTION(node)

Rule 9 — Sinh action hợp lệ

IF position(0) ở góc hoặc cạnh hoặc giữa
THEN
    sinh các action hợp lệ:
    {UP, DOWN, LEFT, RIGHT}

Rule 10 — Tạo child node

IF action hợp lệ
THEN
    child ← CHILD-NODE(problem, node, action)

Rule 11 — Kiểm tra child

IF child.STATE ∉ reached
AND child ∉ frontier
THEN
    frontier.INSERT(child)
    
Rule 12 — Thất bại

IF frontier = EMPTY
THEN
    return failure

P — Performance Measure (Độ đo hiệu suất)
Mục tiêu của agent là tìm được trạng thái đích từ trạng thái ban đầu với số bước di chuyển ít nhất. Hiệu suất được đánh giá dựa trên:

S — Sensors (Bộ phận cảm nhận)
Agent quan sát:
Toàn bộ trạng thái hiện tại của ma trận
Vị trí của ô trống 0
Các hướng di chuyển hợp lệ
Trạng thái goal để kiểm tra đã đạt đích hay chưa

```python
from collections import deque

# START VÀ GOAL

start = "123456807"
goal  = "123456780"

# giới hạn số bước duyệt
LIMIT = 20

# IN MA TRẬN

def print_board(state):

    for i in range(0, 9, 3):
        print(state[i:i+3])

    print()

# SINH ACTION HỢP LỆ

def get_neighbors(state):

    neighbors = []

    zero = state.index('0')

    row = zero // 3
    col = zero % 3

    directions = [
        (-1, 0, "UP"),
        (1, 0, "DOWN"),
        (0, -1, "LEFT"),
        (0, 1, "RIGHT")
    ]

    for dr, dc, action in directions:

        nr = row + dr
        nc = col + dc

        # kiểm tra còn nằm trong bảng
        if 0 <= nr < 3 and 0 <= nc < 3:

            new_pos = nr * 3 + nc

            board = list(state)

            # đổi chỗ ô trống
            board[zero], board[new_pos] = board[new_pos], board[zero]

            neighbors.append(("".join(board), action))

    return neighbors

# BFS

def bfs(start, goal):

    # Rule 1 — Khởi tạo
    node = start

    # Rule 2 — Kiểm tra goal ban đầu
    if node == goal:
        return []

    # Rule 3 — Tạo frontier
    frontier = deque()

    frontier.append((node, []))

    # Rule 4 — Khởi tạo reached
    reached = set()

    step = 0

    # Rule 5 — Kiểm tra frontier
    while frontier and step < LIMIT:

        print("================================")
        print("BUOC:", step + 1)

        print("\nFRONTIER:")

        for item in frontier:
            print(item[0], item[1])

        # Rule 6 — Lấy node đầu tiên
        state, path = frontier.popleft()

        print("\nNODE DANG XET:")
        print_board(state)

        # Rule 7 — Đưa state vào reached
        reached.add(state)

        print("REACHED:")
        print(reached)

        # Rule 8 — Kiểm tra goal
        if state == goal:

            print("TIM THAY GOAL!")
            return path

        print("\nCAC NODE CON:")

        # Rule 9 + 10
        for child, action in get_neighbors(state):

            print("ACTION:", action)
            print_board(child)

            # Rule 11 — Kiểm tra child
            if child not in reached:

                frontier.append((child, path + [action]))

        step += 1

    # Rule 12 — Failure
    return None

# CHẠY

solution = bfs(start, goal)

print("================================")

if solution:

    print("LOI GIAI:")
    print(solution)

else:

    print("KHONG TIM THAY LOI GIAI")
    print("DA DAT GIOI HAN", LIMIT, "BUOC")
```
