# Vacuum Cleaner Agent — PEAS Model
 P — Performance Measure
- Làm sạch càng nhiều càng tốt
- Di chuyển hợp lý
- Giảm số bước di chuyển
 E — Environment
- Môi trường dạng lưới 3x3
- Ô sạch hoặc có bụi
- Robot hoạt động trong phòng
 A — Actuators
- Move Up
- Move Down
- Move Left
- Move Right
- Suck
 S — Sensors
- Cảm biến bụi
- Cảm biến vị trí

# Rule Set — Simple Reflex Vacuum Agent
- Ma trận m x n
- 0 → không có bụi
- 1 → có bụi
- 2 → vị trí robot

Rule 1 — Hút bụi
IF state_value = 1
THEN:
- hút bụi
- state_value = 0

Rule 2 — Di chuyển
Robot chỉ đi các hướng hợp lệ.

Góc trên trái (0,0)
Action = RANDOM(DOWN, RIGHT)

Góc trên phải (0,n-1)
Action = RANDOM(DOWN, LEFT)

Góc dưới trái (m-1,0)
Action = RANDOM(UP, RIGHT)

Góc dưới phải (m-1,n-1)
Action = RANDOM(UP, LEFT)

Cạnh trên
Action = RANDOM(DOWN, LEFT, RIGHT)

Cạnh dưới
Action = RANDOM(UP, LEFT, RIGHT)

Cạnh trái
Action = RANDOM(UP, DOWN, RIGHT)

Cạnh phải
Action = RANDOM(UP, DOWN, LEFT)

Ô giữa
Action = RANDOM(UP, DOWN, LEFT, RIGHT)

```python
import random
import time

# Kích thước ma trận
m = 3
n = 3

# Tạo môi trường ngẫu nhiên
room = [[random.choice([0,1]) for _ in range(n)] for _ in range(m)]

# Vị trí robot ban đầu
x = random.randint(0, m - 1)
y = random.randint(0, n - 1)

# Hàm in ma trận
def print_room():
    for i in range(m):
        for j in range(n):
            # Hiển thị robot
            if i == x and j == y:
                print("2", end=" ")
            else:
                print(room[i][j], end=" ")
        print()
    print()

# Hàm kiểm tra còn bụi không
def has_dust():

    for row in room:
        if 1 in row:
            return True

    return False

# Hàm tìm hướng đi hợp lệ
def possible_moves(x, y):

    moves = []

    if x > 0:
        moves.append("U")

    if x < m - 1:
        moves.append("D")

    if y > 0:
        moves.append("L")

    if y < n - 1:
        moves.append("R")

    return moves

# Bắt đầu mô phỏng
step = 1

while has_dust():

    print(f"===== STEP {step} =====")

    print_room()

    # Nếu ô hiện tại có bụi
    if room[x][y] == 1:

        print(f"Robot hút bụi tại ({x},{y})")

        room[x][y] = 0

    # Chọn hướng đi
    moves = possible_moves(x, y)

    action = random.choice(moves)

    print("Possible moves:", moves)
    print("Robot move:", action)

    # Cập nhật vị trí robot
    if action == "U":
        x -= 1

    elif action == "D":
        x += 1

    elif action == "L":
        y -= 1

    elif action == "R":
        y += 1

    step += 1

    time.sleep(1)

# Làm sạch xong
print("===== HOÀN THÀNH =====")

print("Ma trận sau khi hút sạch:")

# In ma trận cuối cùng
for row in room:
    print(*row)
```
