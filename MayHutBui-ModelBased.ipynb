# Vacuum Cleaner Agent — PEAS Model (Model-Based Agent)
P — Performance Measure
- Làm sạch toàn bộ môi trường
- Giảm số bước di chuyển không cần thiết
- Ưu tiên đi đến ô chưa được khám phá
- Hoàn thành công việc trong thời gian ngắn
E — Environment
- Môi trường dạng lưới 3x3
- Mỗi ô có thể:
+ sạch (0)
+ có bụi (1)
- Robot hoạt động trong phòng kín
- Môi trường thay đổi sau khi robot hút bụi
- Robot có bộ nhớ lưu các ô đã đi qua
A — Actuators
- Move Up
- Move Down
- Move Left
- Move Right
- Suck
S — Sensors
- Cảm biến phát hiện bụi tại ô hiện tại
- Cảm biến xác định vị trí robot
- Bộ nhớ ghi nhận các ô đã đi qua (visited map)
- Cảm biến kiểm tra hướng di chuyển hợp lệ

# Rule Set — Model-Based Vacuum Agent

- Ma trận m x n
- 0 → không có bụi
- 1 → có bụi
- 2 → vị trí robot
- visited = bộ nhớ các ô đã đi
  - 0 → chưa đi qua
  - 1 → đã đi qua

Rule 1 — Hút bụi

IF state_value = 1
THEN:
- hút bụi
- state_value = 0

Rule 2 — Ghi nhớ trạng thái

Sau mỗi lần di chuyển:
visited[x][y] = 1

Robot lưu lại các ô đã đi qua.

Rule 3 — Di chuyển

Robot chỉ đi các hướng hợp lệ.

IF tồn tại ô chưa đi qua
THEN:
- ưu tiên đi ô chưa đi

ELSE:
- chọn ngẫu nhiên trong các hướng hợp lệ

Rule 4 — Các hướng hợp lệ

Góc trên trái (0,0)
Possible = (DOWN, RIGHT)

Góc trên phải (0,n-1)
Possible = (DOWN, LEFT)

Góc dưới trái (m-1,0)
Possible = (UP, RIGHT)

Góc dưới phải (m-1,n-1)
Possible = (UP, LEFT)

Cạnh trên
Possible = (DOWN, LEFT, RIGHT)

Cạnh dưới
Possible = (UP, LEFT, RIGHT)

Cạnh trái
Possible = (UP, DOWN, RIGHT)

Cạnh phải
Possible = (UP, DOWN, LEFT)

Ô giữa
Possible = (UP, DOWN, LEFT, RIGHT)

```python
import random
import time

# Kích thước ma trận
m = 3
n = 3

# Tạo môi trường ngẫu nhiên
room = [[random.choice([0,1]) for _ in range(n)] for _ in range(m)]

# Bộ nhớ của robot
visited = [[0 for _ in range(n)] for _ in range(m)]

# Vị trí robot ban đầu
x = random.randint(0, m - 1)
y = random.randint(0, n - 1)

# Đánh dấu ô đầu tiên đã đi qua
visited[x][y] = 1


# Hàm in ma trận
def print_room():

    for i in range(m):

        for j in range(n):

            # Robot
            if i == x and j == y:
                print("2", end=" ")

            else:
                print(room[i][j], end=" ")

        print()

    print()


# Hàm in bộ nhớ robot
def print_memory():

    print("Visited map:")

    for row in visited:
        print(*row)

    print()


# Kiểm tra còn bụi không
def has_dust():

    for row in room:

        if 1 in row:
            return True

    return False


# Tìm các hướng đi hợp lệ
def possible_moves(x, y):

    moves = []

    if x > 0:
        moves.append(("U", x - 1, y))

    if x < m - 1:
        moves.append(("D", x + 1, y))

    if y > 0:
        moves.append(("L", x, y - 1))

    if y < n - 1:
        moves.append(("R", x, y + 1))

    return moves


# Bắt đầu mô phỏng
step = 1

while has_dust():

    print(f"===== STEP {step} =====")

    print_room()

    print_memory()

    # Nếu ô hiện tại có bụi
    if room[x][y] == 1:

        print(f"Robot hút bụi tại ({x},{y})")

        room[x][y] = 0

    # Lấy các hướng đi
    moves = possible_moves(x, y)

    # Ưu tiên ô chưa đi qua
    unvisited_moves = []

    for move in moves:

        direction, nx, ny = move

        if visited[nx][ny] == 0:
            unvisited_moves.append(move)

    # Nếu còn ô chưa đi → ưu tiên đi
    if unvisited_moves:
        action, nx, ny = random.choice(unvisited_moves)

    # Nếu đi hết rồi → đi ngẫu nhiên
    else:
        action, nx, ny = random.choice(moves)

    print("Robot move:", action)

    # Cập nhật vị trí
    x = nx
    y = ny

    # Ghi nhớ ô đã đi
    visited[x][y] = 1

    step += 1

    time.sleep(1)

# Hoàn thành
print("===== HOÀN THÀNH =====")

print("Ma trận sau khi hút sạch:")

for row in room:
    print(*row)
```
