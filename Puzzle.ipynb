Mô hình PEAS cho bài toán 8-Puzzle
- P (Performance): Số bước di chuyển ít nhất để đạt trạng thái đích, thời gian giải quyết nhanh, chi phí di chuyển thấp.
- E (Environment): Khung ô vuông 3 x  chứa các ô số từ 1 đến 8 và một ô trống (kí hiệu là 0 hoặc _).
- A (Actuators): Các hành động di chuyển ô trống: Sang trái (Left), sang phải (Right), lên trên (Up), xuống dưới (Down).
- S (Sensors): Cảm biến vị trí hiện tại của các con số trong ma trận, nhận diện trạng thái đích.

THIẾT LẬP PERCEPT VÀ RULES
Percept (Nhận thức)
- Ma trận 3 x 3 với các giá trị từ 1 -> 8 và số 0 đại diện cho ô trống.
- Ví dụ trạng thái đích (Goal):
                1 2 3
                4 5 6
                7 8 0
Rules (Quy tắc di chuyển)
Giả sử ô trống nằm tại vị trí (x, y):
- Nếu x > 0: Có thể di chuyển UP (đổi chỗ ô trống với ô phía trên).
- Nếu x < 2: Có thể di chuyển DOWN (đổi chỗ ô trống với ô phía dưới).
- Nếu y > 0: Có thể di chuyển LEFT (đổi chỗ ô trống với ô bên trái).
- Nếu y < 2: Có thể di chuyển RIGHT (đổi chỗ ô trống với ô bên phải).

```python
import random
import time

class Reflex8Puzzle:
    def __init__(self, start_state):
        self.board = [row[:] for row in start_state]
        self.target = [
            [1, 2, 3],
            [4, 5, 6],
            [7, 8, 0]
        ]
        self.size = 3

    def locate_empty(self):
        for r in range(self.size):
            for c in range(self.size):
                if self.board[r][c] == 0:
                    return r, c

    def check_goal(self):
        return self.board == self.target

    def manhattan_distance(self, r, c, number):
        if number == 0:
            return 0

        goal_r = (number - 1) // 3
        goal_c = (number - 1) % 3

        return abs(r - goal_r) + abs(c - goal_c)

    def calculate_score(self, state):
        total = 0

        for r in range(self.size):
            for c in range(self.size):
                total += self.manhattan_distance(r, c, state[r][c])

        return total

    def print_board(self, step):
        print(f"\nSTEP {step}")

        for row in self.board:
            print(" ".join("_" if x == 0 else str(x) for x in row))

        print("=" * 20)

    def choose_action(self):
        r, c = self.locate_empty()

        directions = []

        if r > 0:
            directions.append((-1, 0, "UP"))

        if r < 2:
            directions.append((1, 0, "DOWN"))

        if c > 0:
            directions.append((0, -1, "LEFT"))

        if c < 2:
            directions.append((0, 1, "RIGHT"))

        best_choice = None
        best_value = float("inf")

        for dr, dc, move in directions:
            nr, nc = r + dr, c + dc

            self.board[r][c], self.board[nr][nc] = self.board[nr][nc], self.board[r][c]

            current_score = self.calculate_score(self.board)

            if current_score < best_value:
                best_value = current_score
                best_choice = (nr, nc, move)

            elif current_score == best_value:
                if random.randint(0, 1):
                    best_choice = (nr, nc, move)

            self.board[r][c], self.board[nr][nc] = self.board[nr][nc], self.board[r][c]

        nr, nc, action = best_choice

        self.board[r][c], self.board[nr][nc] = self.board[nr][nc], self.board[r][c]

        return action


if __name__ == "__main__":

    puzzle = [
        [1, 2, 3],
        [5, 6, 4],
        [8, 7, 0]
    ]

    agent = Reflex8Puzzle(puzzle)

    steps = 0
    max_steps = 200

    while not agent.check_goal() and steps < max_steps:
        steps += 1

        agent.print_board(steps)

        move = agent.choose_action()

        print(f"--> Agent chọn: {move}")

        time.sleep(0.2)

    if agent.check_goal():
        agent.print_board(steps + 1)
        print(f"HOÀN THÀNH sau {steps} bước!")

    else:
        print("Không tìm được lời giải, agent bị lặp.")
```
